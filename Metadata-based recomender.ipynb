```python
df = pd.read_csv("movies_metadata.csv")
```

```python
df.head()
```

```python
#Functions to convert all non-integers ID's to NAN
def clean_ids(X):
    try:
        return int(X)
    except:
        return np.nan
    
#Clean the ids of df
df['id'] = df['id'].apply(clean_ids)

#Filter all rows that have a null ID
df = df[df['id'].notnull()]
```

```python
#Convert ID's into integerss
df['id'] = df['id'].astype('int')
key_df['id'] = key_df['id'].astype('int')
cred_df['id'] = cred_df['id'].astype('int')
```

```python
#Merge keywords and credits into your main metadata dataframe
df = df.merge(cred_df, on='id')
df = df.merge(key_df, on='id')
```

```python
df2 = pd.read_csv("movies_metadata_clean.csv")
```

```python
df2['overview'] = df['overview']
df2["id"] = df['id']
df2["cast"] = df['cast']
df2["crew"] = df["crew"]
df2['keywords'] = df['keywords']
```

### Wrangling keywords, cast, and crew

```python
#Convert the stringified objects into the native python objects
from ast import literal_eval

features = ["cast", "crew", "keywords", "genres"]
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
#Print the first cast member of the first movie in df
df2.iloc[0]['crew'][0]
```

```python
#Extract the director name. If director is not listed, return NaN
def get_director(x):
    for crew_member in x:
        if crew_member['job'] == "Director":
            return crew_member['name']
    return np.nan
```

```python
df2["director"] = df2['crew'].apply(get_director)
```

```python
df2.director
```

```python
#Returns the list top 3 elements or entire list; whichever is more
def generate_list(x):
    if isinstance(x, list):
        names = [ele['name'] for ele in x]
        #Check if more than 3 elements exist. If yes return only first three
        #If no, return entire list
        if len(names) > 1:
            names = names.iloc[:3]
        return names
    # return empty list in case of missing/malformed data
    return []
```

```python
#Apply the generate_list function to cast and keywords
df2['cast'] = df['cast'].apply(generate_list)
df2['keywords'] = df['keywords'].apply(generate_list)

#Only consider a maximum of 3 generes
df2['genres'] = df['genres'].apply(lambda x: x[:3])
```
